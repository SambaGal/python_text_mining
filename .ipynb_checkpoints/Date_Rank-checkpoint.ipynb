# Assignment 1

In this assignment, you'll be working with messy medical data and using regex to extract relevant infromation from the data. 

Each line of the `dates.txt` file corresponds to a medical note. Each note has a date that needs to be extracted, but each date is encoded in one of many formats.

The goal of this assignment is to correctly identify all of the different date variants encoded in this dataset and to properly normalize and sort the dates. 

Here is a list of some of the variants you might encounter in this dataset:
* 04/20/2009; 04/20/09; 4/20/09; 4/3/09
* Mar-20-2009; Mar 20, 2009; March 20, 2009;  Mar. 20, 2009; Mar 20 2009;
* 20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
* Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
* Feb 2009; Sep 2009; Oct 2010
* 6/2008; 12/2009
* 2009; 2010

Once you have extracted these date patterns from the text, the next step is to sort them in ascending chronological order accoring to the following rules:
* Assume all dates in xx/xx/xx format are mm/dd/yy
* Assume all dates where year is encoded in only two digits are years from the 1900's (e.g. 1/5/89 is January 5th, 1989)
* If the day is missing (e.g. 9/2009), assume it is the first day of the month (e.g. September 1, 2009).
* If the month is missing (e.g. 2010), assume it is the first of January of that year (e.g. January 1, 2010).
* Watch out for potential typos as this is a raw, real-life derived dataset.

With these rules in mind, find the correct date in each note and return a pandas Series in chronological order of the original Series' indices.

For example if the original series was this:

    0    1999
    1    2010
    2    1978
    3    2015
    4    1985

Your function should return this:

    0    2
    1    4
    2    0
    3    1
    4    3

Your score will be calculated using [Kendall's tau](https://en.wikipedia.org/wiki/Kendall_rank_correlation_coefficient), a correlation measure for ordinal data.

*This function should return a Series of length 500 and dtype int.*

```python
import pandas as pd
import re
doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

df = pd.Series(doc)
df.head(10)
```

```python
#pd.set_option('display.max_colwidth', -1)
#pd.set_option("display.max_rows", None, "display.max_columns", None)


```

```python
def date_sorter():
    
    #create search and capture patterns
    pattern1 = r'\d{1,2}[/-]\d{1,2}[/-]\d{2,4}'
    pattern1 = re.compile(pattern1)
    capture1 = r'(\d{1,2}[/-]\d{1,2}[/-]\d{2,4})'


    pattern2 = r'(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*\.?[-\s]\d{1,2},?[-\s]\d{2,4}'
    pattern2 = re.compile(pattern2)
    capture2 = r'((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*\.?[-\s]\d{1,2},?[-\s]\d{2,4})'


    pattern3 = r'\d{1,2}\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{2,4}'
    pattern3 = re.compile(pattern3)
    capture3 = r'(\d{1,2}\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{2,4})'



    #pattern4 = r'\d{1,2}(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{4}'
    #pattern4 = re.compile(pattern4)
    #capture4 = r'(\d{1,2}(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{4})'


    pattern5 = r'(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{4}'
    pattern5 = re.compile(pattern5)
    capture5 = r'((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[\.,]?\s\d{4})'


    pattern6 = r'(?<!\d)(?<!/)\d{1,2}[/-]\d{4}'
    pattern6 = re.compile(pattern6)
    capture6 = r'((?<!\d)(?<!/)\d{1,2}[/-]\d{4})'


    pattern7 = re.compile(r'\b\w?\d{4}\b')
    capture7 = r'\b\w?(\d{4}\b)' 

    #create a data frame and add pattern/raw_date columns
    df_dates = pd.DataFrame(df, columns = ['string'])


    df_dates['pattern'] = ''

    df_dates['raw_date'] = ''

    #search for patterns, create column labeling each pattern
    for i in range(len(df_dates)):
        if bool(pattern1.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat1'
            df_dates['raw_date'][i] = re.search(capture1,df_dates['string'][i]).group(1)
        elif bool(pattern2.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat2'
            df_dates['raw_date'][i] = re.search(capture2,df_dates['string'][i]).group(1)
        elif bool(pattern3.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat3'
            df_dates['raw_date'][i] = re.search(capture3,df_dates['string'][i]).group(1)
        #elif bool(pattern4.search(df_dates['string'][i])):
            #df_dates['pattern'][i] = 'pat4'
            #df_dates['raw_date'][i] = re.search(capture4,df_dates['string'][i]).group(1)
        elif bool(pattern5.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat5'
            df_dates['raw_date'][i] = re.search(capture5,df_dates['string'][i]).group(1)
        elif bool(pattern6.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat6'
            df_dates['raw_date'][i] = re.search(capture6,df_dates['string'][i]).group(1)
        elif bool(pattern7.search(df_dates['string'][i])):
            df_dates['pattern'][i] = 'pat7'
            df_dates['raw_date'][i] = re.search(capture7,df_dates['string'][i]).group(1)
        else:
            df_dates['pattern'][i] = 'needs pat'

    #replace dashes
    df_dates['raw_date'] = df_dates['raw_date'].str.replace('-', '/')

    #remove dots and commas
    df_dates['raw_date'] = df_dates['raw_date'].str.replace('[\.,]','')

    #create month map, include misspelled months
    month_map = {'Jan': '01', 'Feb': '02', 'Mar':'03', 'Apr': '04', 'May': '05', 'Jun': '06',
            'Jul': '07', 'Aug':'08', 'Sep':'09', 'Oct': '10', 'Nov': '11', 'Dec':'12',
             'January': '01', 'February': '02', 'March':'03', 'April': '04', 'May': '05', 'June': '06',
            'July': '07', 'August':'08', 'September':'09', 'October': '10', 
             'November': '11', 'December':'12','Decemeber': '12', 'Janaury': '01', '2June': '06'}

    #add month and day to incomplete dates
    for i in range(len(df_dates)):
        if df_dates['pattern'][i]=='pat5':
            df_dates['raw_date'][i] = '01 ' + df_dates['raw_date'][i]
        #elif df_dates['pattern'][i]=='pat4':
            #df_dates['raw_date'][i]='02 ' + df_dates['raw_date'][i]
        elif df_dates['pattern'][i]=='pat6':
            df_dates['raw_date'][i] = '01/' + df_dates['raw_date'][i]
        elif df_dates['pattern'][i]=='pat7':
            df_dates['raw_date'][i] = '01/01/' + df_dates['raw_date'][i]
        else:
            df_dates['raw_date'][i] = df_dates['raw_date'][i]

    #replace / with space
    df_dates['raw_date'] = df_dates['raw_date'].str.replace('/',' ')

    #split into three columns
    new = df_dates['raw_date'].str.split(' ', n=2, expand=True )
    df_dates['month'] = new[0]
    df_dates['day'] = new[1]
    df_dates['year'] = new[2]

    #swap day/month for pat3, pat5 , pat 4 and pat6
    df_dates[['month','day']] = df_dates[['day','month']].where(df_dates['pattern'] == 'pat5', df_dates[['month','day']].values)
    df_dates[['month','day']] = df_dates[['day','month']].where(df_dates['pattern'] == 'pat6', df_dates[['month','day']].values)
    df_dates[['month','day']] = df_dates[['day','month']].where(df_dates['pattern'] == 'pat3', df_dates[['month','day']].values)
    #df_dates[['month','day']] = df_dates[['day','month']].where(df_dates['pattern'] == 'pat4', df_dates[['month','day']].values)

    #add 19 to 2 digit year
    df_dates['year'] = df_dates['year'].apply(lambda x: '19'+x if len(x)==2 else x)

    #add 0 to 1 digit day
    df_dates['day']=df_dates['day'].apply(lambda x: '0'+x if len(x)==1 else x)

    #add 0 to 1 digit month
    df_dates['month'] = df_dates['month'].apply(lambda x: '0'+x if len(x)==1 else x)

    #replace 'letter' months
    df_dates.replace({'month': month_map}, inplace=True)

    #create final date column
    df_dates['date'] = df_dates['month']+'/'+df_dates['day']+'/'+df_dates['year']

    #convert to datetime
    df_dates['date'] = pd.to_datetime(df_dates['date'], format='%m/%d/%Y')

    #create rank column
    df_dates['rank'] = df_dates.sort_values(by='date').index
  

    return df_dates['rank']
```
